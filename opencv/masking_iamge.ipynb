```python
import cv2 as cv

import numpy as np

from PIL import Image

im_path = "./pic/baboon.png"

im_array = cv.imread(im_path)

display(Image.fromarray(cv.cvtColor(im_array, cv.COLOR_BGR2RGB)))

#creating a mask 
mask = im_array[:,:,2] > 201
#apply the mask to the image by setting the corresponding pixels to black
im_array[mask] = [0,0,0]

display(Image.fromarray(cv.cvtColor(im_array, cv.COLOR_BGR2RGB)))
```
