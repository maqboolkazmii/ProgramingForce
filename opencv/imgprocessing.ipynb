```python
import cv2

import numpy as np

from PIL import Image

im_path = "./pic/baboon.png"

im_array = cv2.imread(im_path)

display(Image.fromarray(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)))


#converting BGR to HSV
img_hsv=cv2.cvtColor(im_array, cv2.COLOR_BGR2HSV)

# to get more accurate pixels we creating two mask

# lower mask (0-10)
lower_red = np.array([0,50,50])
upper_red = np.array([10,255,255])
mask0 = cv2.inRange(img_hsv, lower_red, upper_red)
  
# upper mask (170-180)
lower_red = np.array([170,50,50])
upper_red = np.array([180,255,255])
mask1 = cv2.inRange(img_hsv, lower_red, upper_red)

# join my masks
mask = mask0+mask1

im_array[mask == 255] =0

display(Image.fromarray(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)))
```
